```python
import random
import csv

with open("files/netflix_titles.csv", "r") as f:
    csv_reader = csv.reader(f)  # Create a CSV reader

    next(csv_reader)  # Skip the header line

    movies = list(csv_reader)  # Read the CSV data into a list

    # Randomly select a movie from the list
    random_movie = random.choice(movies)

    # Extract the desired information from the movie data
    key = random_movie[0]
    type = random_movie[1]
    title = random_movie[2]
    listed_in = random_movie[10]
    rating = random_movie[8]
    description = random_movie[11]

    # Print the randomly generated movie details
    print("Type:", type)
    print("Title:", title)
    print("Listed In:", listed_in)
    print("Rating:", rating)
    print("Description:", description)
```

```python
import requests
import random

url = "https://netflix54.p.rapidapi.com/search/"

querystring = {
    "query": "stranger",
    "offset": "0",
    "limit_titles": "50",
    "limit_suggestions": "20",
    "lang": "en"
}

headers = {
    "X-RapidAPI-Key": "c6f189e0ebmshda971f52a1a73a2p11b953jsn1a3dac83ec45",
    "X-RapidAPI-Host": "netflix54.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

movies = response.json()

# Extract the titles from the JSON response
if "titles" in movies:
    titles = [movie["title"] for movie in movies["titles"]]
    if titles:
        random_movie = random.choice(titles)
        print("Random Movie:", random_movie)
    else:
        print("No movie titles found.")
else:
    print("Error: Failed to fetch data from the API.")
```
